```
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-dark')

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix
```

```
iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

lda2 = LinearDiscriminantAnalysis(n_components=2, tol=1.0e-4)
X_r2 = lda2.fit(X, y).transform(X)

plt.figure(figsize=(6,4), dpi=200)
colors = ['navy', 'turquoise', 'darkorange']

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
  plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.5, color=color,label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')
plt.xlim(-10.5,10.5)
plt.ylim(-3,3)
plt.show()


def myLDA(index, X=X, y=y):

  X = X[y != index]
  y = y[y != index]

  lda = LinearDiscriminantAnalysis(n_components=1,tol=1.0e-4)
  X_r1 = lda.fit(X, y).transform(X)
    
  plt.figure(figsize=(4,2), dpi=200)
  colors = ['navy', 'turquoise', 'darkorange']
  for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r1[y == i, 0], np.full(len(X_r1[y == i, 0]),i/5.), s=15, alpha=.5, color=color, label=target_name)
  plt.legend(loc='best', shadow=False, scatterpoints=1)
  plt.title('LDA of IRIS dataset (%s is out)' % target_names[index], fontsize=10)
  
  if index == 0: 
    plt.axvline(x=0.17,linewidth=0.2, color='red')
    #for i in range(len(X_r1)): print(X_r1[i],y[i])
      
  plt.ylim(-2,4)
  plt.yticks([])
  plt.show()
  
  return lda, X, y

threshold = 0.72
for i in range(0,3): 
  lda, X, Y_true = myLDA(i)
  Y_pred = lda.predict(X)
  print(confusion_matrix(Y_true, Y_pred))

  #print(list(lda.predict_proba(X)[:,1] > threshold).count(True)) # https://web.stanford.edu/class/stats202/content/lec9.pdf page 18
  for threshold in range(57,58): #(27,44) for 1: #(57,74) for 2:
    threshold = threshold/100.
    Y_pred = lda.predict_proba(X)[:,1] > threshold
    print("threshold = %f" % threshold)
    print(confusion_matrix(Y_true, Y_pred))
  print(lda.predict_proba(X))
```
